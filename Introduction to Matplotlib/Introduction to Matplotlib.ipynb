## Mathplotlib example workflow


```python
# 0. import mathplotlib and get it ready for plotting in jupyter
%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare data 
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig , ax = plt.subplots(figsize=(10,8)) # figsize(width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simplt plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("./sample-plot.png")
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2)
ax1.set(title="Plot")
ax2.bar(x, np.exp(x))
ax2.set(title="Bar")
ax3.scatter(np.random.randn(10), np.random.randn(10))
ax3.set(title="Scatter")
ax4.hist(np.random.rand(10))
ax4.set(title="Histogram")
ax5.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax5.set(title="bar")
ax6.hist(x)
ax6.set(title="histogram")
```

### Which one should you use? (pyplot vs matplotlib 00 methods)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the oo method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50
```

```python
fig , ax = plt.subplots()
ax.scatter(over_50["age"], over_50["chol"])
```

```python
fig , ax = plt.subplots()
ax.scatter(x=over_50["age"], y=over_50["trestbps"], c = over_50["target"])
```

```python
# oo methods 
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)
ax.set_xlim([50, 80])
```

```python
# With the pyplot
over_50.plot(kind="scatter", x='age', y='chol', c='target', figsize=(10,6))
```

```python
# Object oriented methods from scrach
fig, ax= plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c = over_50["target"])

# Costomize the plot
ax.set(title="Heart Disease and Cholosterl Levels", xlabel="age", ylabel="Chol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="target")

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
fig, ((ax1, ax2)) = plt.subplots(nrows= 2, ncols= 1, figsize=(10,10), sharex=True)

# For 1st figure
scatter = ax1.scatter(y=over_50["chol"], x=over_50["age"],c = over_50['target'])
ax1.set(ylabel='chol', title="Heart Disease and Cholestrol Graph")
ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(300)

# For 2nd figure
scatter = ax2.scatter(y=over_50["thalach"], x=over_50["age"],c = over_50['target'])
ax2.set(xlabel='age', ylabel='thalach', title="Heart Disease and thalach Graph")
ax2.legend(*scatter.legend_elements(), title='target')
ax2.axhline(over_50['thalach'].mean(), linestyle='--')


# Add the title to the figure
fig.suptitle("Heart Disease and chol", fontsize=16, fontweight=10)
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# Object oriented methods from scrach
fig, ax= plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c = over_50["target"],
                    cmap='winter') # All this means this cchanges the color skin winter, viridis, plasma

# Costomize the plot
ax.set(title="Heart Disease and Cholosterl Levels", xlabel="age", ylabel="Chol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="target")

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# Customizing the y and x axis

fig, ((ax1, ax2)) = plt.subplots(nrows= 2, ncols= 1, figsize=(10,10), sharex=True)

# For 1st figure
scatter = ax1.scatter(y=over_50["chol"], x=over_50["age"],c = over_50['target'], cmap='winter')
ax1.set(ylabel='chol', title="Heart Disease and Cholestrol Graph")
# Change the x axis limits
ax1.set_xlim([50, 80])

ax1.legend(*scatter.legend_elements(), title='target')
ax1.axhline(300)

# For 2nd figure
scatter = ax2.scatter(y=over_50["thalach"], x=over_50["age"],c = over_50['target'], cmap='winter')
ax2.set(xlabel='age', ylabel='thalach', title="Heart Disease and thalach Graph")
# Change the y-axis limits
ax2.set_ylim([60, 200])
ax2.legend(*scatter.legend_elements(), title='target')
ax2.axhline(over_50['thalach'].mean(), linestyle='--')


# Add the title to the figure
fig.suptitle("Heart Disease and chol", fontsize=16, fontweight=10)
```
